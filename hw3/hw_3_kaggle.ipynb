```python
from sklearn.linear_model import LogisticRegression, LogisticRegressionCV
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler, RobustScaler, LabelEncoder, OneHotEncoder
from sklearn.pipeline import make_pipeline
```

```python
data.head()
```

```python
#     target
X_df = data.drop(target_col, axis=1)
y_df = data[target_col]

X_df_train, X_df_test, y_df_train, y_df_test = train_test_split(X_df, y_df,
                                                                train_size=0.8,
                                                                random_state=42)
```

```python
from sklearn.compose import ColumnTransformer
```

```python
#    
#     ,      
# 
encoder = ColumnTransformer([
        ('num', RobustScaler(), num_cols),
        ('cat', OneHotEncoder(drop='first'), cat_cols)
    ], remainder='passthrough')

X_train_encoded = encoder.fit_transform(X_df_train)
y_train = y_df_train.values
```

```python
#   
encoder.get_feature_names_out()
```

```python
base_model = LogisticRegressionCV(cv=5, scoring='roc_auc', refit=True)
base_model.fit(X_train_encoded, y_train)

res_model = make_pipeline(encoder, base_model)

print('  C:', base_model.C_)

print('ROC-AUC (train):',
      roc_auc_score(y_df_train, res_model.predict_proba(X_df_train)[:, 1]))
print('ROC-AUC (test):',
      roc_auc_score(y_df_test, res_model.predict_proba(X_df_test)[:, 1]))
```

 GridSearchCV:

```python
gscv = GridSearchCV(estimator=LogisticRegression(),
             param_grid={'C': [100.0, 10.0, 1.0, 0.1, 0.01, 0.001]},
             cv=5,
             scoring='roc_auc',
             n_jobs=-1,
             verbose=10)

gscv.fit(X_train_encoded, y_train)

print('  C:', gscv.best_params_['C'])

res_model_gs = make_pipeline(encoder, gscv)

print('ROC-AUC (train):',
      roc_auc_score(y_df_train, res_model_gs.predict_proba(X_df_train)[:, 1]))
print('ROC-AUC (test):',
      roc_auc_score(y_df_test, res_model_gs.predict_proba(X_df_test)[:, 1]))
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_df, y_df,
                                                    train_size=0.8,
                                                    random_state=42)

categorical_indices, = data.columns.isin(cat_cols).nonzero()

# boosting_model = catboost.CatBoostClassifier(n_estimators=200,
#                                              cat_features=categorical_indices)
#  
# boosting_model = catboost.CatBoostClassifier(cat_features=categorical_indices)
#  
boosting_model = catboost.CatBoostClassifier(n_estimators=500,
                                             learning_rate = 0.01,
                                             cat_features=categorical_indices)

# n_estimators learnin_rate ROC-AUC_train ROC-AUC_test
# default      default      0.90024845412 0.8237102959
# 100          default      0.87350926038 0.8225984613
# 200          default      0.89306567006 0.8217157466
# 500          default      0.89903341711 0.8241830065 #    test
#                                                      #  
# 1500         default      0.90040302718 0.8229964064
# 500          0.1          0.93832205783 0.8097846276
# 500          0.01         0.87122777395 0.8252200757 #   n = 500
# 500          0.001        0.84963842174 0.8115500567


boosting_model.fit(X_train, y_train)

y_train_predicted = boosting_model.predict_proba(X_train)[:, 1]
y_test_predicted = boosting_model.predict_proba(X_test)[:, 1]

print('ROC-AUC (train)', roc_auc_score(y_train, y_train_predicted))
print('ROC-AUC (test)', roc_auc_score(y_test, y_test_predicted))
```
